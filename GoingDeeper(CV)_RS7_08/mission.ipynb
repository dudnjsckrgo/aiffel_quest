- OBJECT DETECTION 의 METRIC
    - TN / FP / FN 
    TN:객체가 없는 영역을 정확하게 객체가 없다고 예측한 경우의 수

    FP:객체가 없는 영역을 잘못해서 객체가 있다고 예측한 경우의 수

    FN:실제로 객체가 있는데, 모델이 그 객체를 탐지하지 못한 경우의 수

    - Precision / Recall
     Precision: 모델이 객체가 있다고 예측한 경우 중 실제로 객체가 있는 비율.

     Recall: 실제로 객체가 있는 모든 경우중에서 모델이 정확하게 객체가 있다고 예측한 비율
    
    - IoU
    
    IoU :예측된 바운딩 박스와 실제 바운딩 박스의 교집합을 두 박스의 합집합으로 나눈값 
    
    - mAP 모든 클래스에대한 AP 의 평균값 AP는 precision-recall 곡선 아래의 면적을 나타내며, 이는 다양한 임계값에서 모델이 얼마나 잘 성능을 내는지를 나타냄.
